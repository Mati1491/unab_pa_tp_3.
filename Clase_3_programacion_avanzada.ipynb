```python
class Punto:
  def __init__(self, x , y):
    self.x = x
    self.y = y

  def eje_x(self):
    return self.x

  def eje_y(self):
    return self.y

  def impresion (self):
    return f" El punto se encuentra x= {self.x} y = {self.y}"

  def opuesto(self):
    nuevox = self.x * -1
    nuevoy = self.y * -1
    return Punto(nuevox ,nuevoy)
```

```python
class Linea():
  def __init__(self, punto_a ,punto_b):
    self.punto_a = punto_a
    self.punto_b = punto_b

  def mueva_derecha(self, n):
    self.punto_a.x += n
    self.punto_b.x += n

  def mueva_izquierda(self, n):
    self.punto_a.x += n
    self.punto_b.x += n

  def mueva_arriba(self, n):
    self.punto_a.y += n
    self.punto_b.y += n

  def mueva_abajo(self, n):
    self.punto_a.y -= n
    self.punto_b.y -= n

```

```python
punto_a = Punto(8,10)
punto_b = Punto(11,12)
linea = Linea(punto_a, punto_b)
print(linea.punto_a.impresion())
linea.mueva_derecha(2)
print(linea.punto_a.impresion())
```
