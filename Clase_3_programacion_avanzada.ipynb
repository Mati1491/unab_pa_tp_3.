```python
class Punto:
  def __init__(self, x , y):
    self.x = x
    self.y = y

  def eje_x(self):
    return self.x

  def eje_y(self):
    return self.y

  def impresion (self):
    return f" x= ({self.y} y = {self.y})"

  def opuesto(self):
    nuevox = self.x * -1
    nuevoy = self.y * -1
    return Punto(nuevox ,nuevoy)
```

```python
class Linea():
  def __init__(self, punto_a ,punto_b):
    self.punto_a = punto_a
    self.punto_b = punto_b

  def mueva_derecha(self, d):
    self.punto_a.x += d
    self.punto_b.y +=d







```
