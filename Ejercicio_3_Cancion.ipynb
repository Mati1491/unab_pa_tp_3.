```python
class Cancion:
  def __init__(self,titulo, autor):
    self.titulo = titulo
    self.autor = autor

  #metodos
  def get_titulo(self):
    return self.titulo

  def get_autor(self):
    return self.autor

  def set_titulo(self, titulo):
    self.titulo = titulo

  def set_autor(self,autor):
    self.autor = autor

  def mostrar(self):
    return f"El titulo de la cancion es {self.titulo} y el autor es {self.autor}"
```

```python
cancion1 = Cancion("Ruleta", "Los piojos")
print(cancion1.get_autor())
print(cancion1.mostrar())
```
